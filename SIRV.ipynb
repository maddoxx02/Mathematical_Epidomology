# SIRV Model

#### The four Different Groups of People here are:

1. Susecptible 
2. Infected
3. Recovered 
4. Vaccinated 

------------

### Importing Libraries

```python
import numpy as np
import seaborn as sns
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

---------

### Parameters 

```python
N = 1000          # Total Population

# Parameters

BETA = 0.4        # Contact Rate within population 

GAMMA = 0.001     # Recovery Rate within (1/days)  Gamma = 1/days

VAC_R = 0.02      # Vaccination Rate (1/days)


#Intial Conditions

I0 = 3             # INitial number of Infected Individuals
#I0 = (VAC_R * N)  # INitial number of Infected Individuals (SEMI TIME)

R0 = 0            # Initial NUmber of recovered Individuals

S0 = N-I0-V0-R0  # Initial number of Susceptivel Individuals

V0 = 0            # Inital number of Vaccinated Individuals


#Time

t = np.linspace(0,2000,2000) # A Grid of Time Points (in days)
```

-------

## SIRV Differential Function

```python

# The SIRV model differential equations.

def deriv(y, t, N, BETA, GAMMA):
    S, I, R, V = y
    dSdt = -BETA * S * I / N - (VAC_R * S)
    dIdt = BETA * S * I / N - (GAMMA * I) 
    dRdt = GAMMA * I
    dVdt = VAC_R*S
    
    return dSdt, dIdt, dRdt, dVdt
```

-------

### Initial Vector

```python
# Initial conditions vector
y0 = S0, I0, R0, V0

# Integrate the SIR equations over the time grid, t.

ret = odeint(deriv, y0, t, args=(N, BETA, GAMMA))
S, I, R, V = ret.T
```

# Normal Scenario of Disease & Vaccination

```python
N = 1000          # Total Population

# Parameters

BETA = 0.2        # Contact Rate within population 

GAMMA = 0.05    # Recovery Rate within (1/days)  Gamma = 1/days

VAC_R = 0.005      # Vaccination Rate (1/days)


#Intial Conditions
#NU = 0.01

I0 = 1             # INitial number of Infected Individuals (PATIENT ZERO)
#I0 = (NU * N)  # INitial number of Infected Individuals (SEMI TIME)

R0 = 0            # Initial NUmber of recovered Individuals

S0 = N-I0-V0-R0  # Initial number of Susceptivel Individuals
#S0 =(1-NU)*N

V0 = 0            # Inital number of Vaccinated Individuals


#Time

t = np.linspace(0,200,200) # A Grid of Time Points (in days)




# Initial conditions vector
y0 = S0, I0, R0, V0

# Integrate the SIR equations over the time grid, t.

ret = odeint(deriv, y0, t, args=(N, BETA, GAMMA))
S, I, R, V = ret.T


# Plot the data on three separate curves for S(t), I(t) and R(t)


plt.figure(facecolor='w')
plt.figure(figsize = (12,8), dpi = 100)

sns.set(style="darkgrid")

plt.plot(t, S, 'b', lw = 3, label='Susceptible')
plt.plot(t, I, 'r', lw = 3, label='Infected')
plt.plot(t, R, 'g', lw = 3, label='Recovered ')
plt.plot(t, V, 'purple', lw = 3, label='Vaccinated')

plt.xlabel('Time (in days)')
plt.ylabel('Number of People (in 100s)')
plt.title("SIRV Model - Normal Case")

plt.legend()

plt.show()


```
